# **Synth√®se ‚Äî √âquations diff√©rentielles lin√©aires**

## 1. Introduction

Les √©quations diff√©rentielles mod√©lisent les lois de la physique (Newton, Euler, Lagrange).
Exemple fondamental : le **mouvement d‚Äôun point mat√©riel** :
[
m \vec{x}''(t) = \vec{F}(t)
]

### Forces et √©quations associ√©es :

| Type de force                           | √âquation diff√©rentielle                                           | Ordre |
| --------------------------------------- | ----------------------------------------------------------------- | ----- |
| Force externe seule                     | ( m\vec{x}''(t) = \vec{F}(t) )                                    | 2     |
| + frottement proportionnel √† la vitesse | ( \vec{F}(t) - \gamma \vec{x}'(t) = m\vec{x}''(t) )               | 2     |
| + ressort (force √©lastique)             | ( \vec{F}(t) - \gamma \vec{x}'(t) - k\vec{x}(t) = m\vec{x}''(t) ) | 2     |

üëâ **√Ä retenir** : les √©quations du mouvement sont lin√©aires tant que les forces d√©pendent *lin√©airement* de la position et de la vitesse.

## 2. √âquations diff√©rentielles lin√©aires du premier ordre

### D√©finition

[
x'(t) = P(t)x(t) + Q(t)
]
avec ( P, Q ) continues.

### Cas homog√®ne (( Q = 0 ))

[
x(t) = x(t_0)e^{\int_{t_0}^{t} P(s),ds}
]

### Cas g√©n√©ral (( Q \neq 0 )) ‚Äî **Formule de Duhamel**

[
x(t) = x(t_0)e^{\int_{t_0}^{t} P(s),ds} + \int_{t_0}^{t} e^{\int_{s}^{t} P(u),du} Q(s),ds
]

### Exemple type

[
x'(t) = x(t) + t,\quad x(0) = x_0
]

**Solution :**
[
x(t) = x_0 e^{t} - t + e^{t} - 1
]
(v√©rifi√© par int√©gration symbolique sous `sympy`).

---

## 3. Syst√®mes d‚Äô√©quations diff√©rentielles lin√©aires du premier ordre

### D√©finition

[
X'(t) = A(t)X(t) + Q(t)
]
o√π ( X ) est un vecteur, ( A(t) ) une matrice de fonctions continues.

### Exemple (m√©canique du point)

En dimension 2 :
[
\begin{cases}
x'_1 = v_1,\
x'_2 = v_2,\
m v'_1 = F_1 - \gamma v_1 - kx_1,\
m v'_2 = F_2 - \gamma v_2 - kx_2
\end{cases}
]

### Exponentielle de matrice

[
e^{A} = \sum_{n=0}^{\infty} \frac{A^n}{n!}
]

### Solution g√©n√©rale (cas (A) constante)

[
X(t) = e^{A(t-t_0)}X(t_0) + \int_{t_0}^{t} e^{A(t-s)}Q(s),ds
]

### Exemple

[
\begin{cases}
x'_1 = x_1 + x_2 + t,\
x'*2 = x_1 - x_2 + t
\end{cases}
\Rightarrow
X(t)=e^{A(t-t_0)}X(t_0)+\int*{t_0}^{t} e^{A(t-s)}\begin{pmatrix}s\s\end{pmatrix}ds
]
avec (A=\begin{pmatrix}1&1\1&-1\end{pmatrix}).

## 4. Approximation num√©rique des solutions

### √âquation de r√©f√©rence

[
x'(t) = f(t,x(t)),\quad x(0)=x_0
]

### Discr√©tisation

( t_n = n\Delta t,\ \Delta t = T/n )

---

### M√©thode d‚ÄôEuler explicite

[
x_{n+1} = x_n + \Delta t f(t_n, x_n)
]

### M√©thode d‚ÄôEuler implicite

[
x_{n+1} = x_n + \Delta t f(t_{n+1}, x_{n+1})
]

### M√©thode de Heun (am√©lior√©e)

[
x_{n+1} = x_n + \frac{\Delta t}{2}\left[f(t_n,x_n) + f(t_{n+1}, x_n + \Delta t f(t_n,x_n))\right]
]

---

### Exemple : ( x'(t)=x(t),~x(0)=1 )

* **Exacte** : ( x(t)=e^t )
* **Euler explicite** : ( x_{n+1}=(1+\Delta t)x_n )
* **Heun** : am√©liore la pr√©cision d‚Äôordre 2.

---

### Stabilit√© asymptotique

#### Cas : ( x'(t) = -\gamma x(t) ), ( \gamma>0 )

| Sch√©ma          | Formule                                     | Condition de stabilit√©  |
| --------------- | ------------------------------------------- | ----------------------- |
| Euler explicite | ( x_{n+1} = (1-\Delta t\gamma)x_n )         | ( \Delta t < 2/\gamma ) |
| Euler implicite | ( x_{n+1} = \frac{1}{1+\Delta t\gamma}x_n ) | stable ‚àÄ ( \Delta t>0 ) |

üëâ Les sch√©mas implicites sont **inconditionnellement stables**.

## 5. Probl√®mes aux limites du second ordre

### D√©finition g√©n√©rale

[
-(\alpha(x)u'(x))' + \beta(x)u'(x) + \gamma(x)u(x) = f(x), \quad x\in[a,b]
]
avec conditions aux limites :

* **Dirichlet** : ( u(a)=u_a,~u(b)=u_b )
* **Neumann** : ( u'(a)=u'_a,~u'(b)=u'_b )
* **Robin** : ( u(a)-\delta_a u'(a)=u_a,\ u(b)+\delta_b u'(b)=u_b )

### Interpr√©tation physique

* ( \alpha ): diffusion
* ( \beta ): convection
* ( \gamma ): amortissement
* ( f ): source

---

### Solution g√©n√©rale (sans condition aux bords)

[
u(x) = c_1 u_1(x) + c_2 u_2(x) + u_p(x)
]
o√π (u_1,u_2) : solutions homog√®nes et (u_p) : solution particuli√®re.

---

### Exemples

#### (a) Pas de solution

[
u''(x)=1,\quad u'(0)=u'(1)=0
]
‚Üí Contradiction sur (u'(1)) ‚áí **pas de solution**.

#### (b) Avec solution

[
u''(x)=1,\quad u(0)=0,~u'(1)=0
]
‚Üí (u(x)=\frac{x^2}{2}-x)

---

### Exercices typiques

1. (u''=1,\ u(0)=u(1)=0)
2. (u''+u'=1,\ u(0)=u(1)=0)
3. (u''+u=1,\ u'(0)=u'(1)=0)

## 6. Discr√©tisation num√©rique d‚Äôun probl√®me 1D

### Maillage

[
x_i = i h, \quad i = 0, \ldots, N+1, \quad h = \frac{1}{N+1}
]

### Sch√©ma de diff√©rences finies (cas (-u''=f))

[
\frac{-u_{i-1}+2u_i-u_{i+1}}{h^2}=f_i,\quad u_0=u_{N+1}=0
]

### Syst√®me matriciel

[
A U = F,\quad A=\frac{1}{h^2}
\begin{pmatrix}
2 & -1 & 0 & \cdots & 0\
-1 & 2 & -1 & \cdots & 0\
0 & -1 & 2 & \ddots & \vdots\
\vdots & \ddots & \ddots & \ddots & -1\
0 & \cdots & 0 & -1 & 2
\end{pmatrix}
]

**Impl√©mentation Python (creuse) :**

```python
from scipy.sparse import diags
import numpy as np

N = 10
h = 1/(N+1)
A = diags([-1, 2, -1], [-1, 0, 1], shape=(N, N)) / h**2
```

## 2. **Syst√®mes lin√©aires du premier ordre**

## 3. **M√©thodes num√©riques (Euler / Heun)**

### üîπ Type d‚Äôexercices probables :

* Appliquer un sch√©ma d‚ÄôEuler explicite √† une √©quation donn√©e
* Comparer Euler explicite / implicite / Heun
* √âtudier la **stabilit√© asymptotique** d‚Äôun sch√©ma
* D√©terminer la **condition sur Œît**

### üîπ Sch√©mas √† conna√Ætre :

| Sch√©ma          | Formule                                                                         | Commentaire                               |
| --------------- | ------------------------------------------------------------------------------- | ----------------------------------------- |
| Euler explicite | (x_{n+1}=x_n+\Delta t,f(t_n,x_n))                                               | simple, instable si Œît trop grand         |
| Euler implicite | (x_{n+1}=x_n+\Delta t,f(t_{n+1},x_{n+1}))                                       | stable pour tout Œît                       |
| Heun            | (x_{n+1}=x_n+\frac{\Delta t}{2}[f(t_n,x_n)+f(t_{n+1},x_n+\Delta t f(t_n,x_n))]) | meilleure pr√©cision, pas de non-lin√©arit√© |

### üîπ Exemple classique :

[
x'(t)=-\gamma x(t),\quad x_0=1
]

| Sch√©ma          | Relation                                 | Stabilit√©                       |
| --------------- | ---------------------------------------- | ------------------------------- |
| Euler explicite | (x_{n+1}=(1-\gamma \Delta t)x_n)         | stable si (0<\Delta t<2/\gamma) |
| Euler implicite | (x_{n+1}=\frac{1}{1+\gamma \Delta t}x_n) | stable ‚àÄ Œît                     |

### üîπ √Ä retenir :

* **Stabilit√© asymptotique = bornitude de la suite**
* Les sch√©mas **implicites** sont inconditionnellement stables.

## 4. **Probl√®mes aux limites du second ordre**

### üîπ Type d‚Äôexercices probables :

* R√©soudre une EDL du second ordre avec conditions aux bords
* V√©rifier si une solution existe (ou pas)
* Donner l‚Äôexpression de la solution g√©n√©rale homog√®ne + particuli√®re
* Discr√©tiser (-u''=f) sur ([0,1])

### üîπ Forme g√©n√©rale :

[
-(\alpha(x)u'(x))' + \beta(x)u'(x) + \gamma(x)u(x) = f(x)
]
avec conditions de Dirichlet, Neumann ou Robin.

### üîπ Forme r√©duite √† conna√Ætre :

[

### üîπ Exemples √† retenir :

   (u''=1,\ u(0)=0,~u'(1)=0 \Rightarrow u(x)=\frac{x^2}{2}-x)

## 5. **Discr√©tisation du segment ([0,1])**

### üîπ Code √† conna√Ætre :
