```python
import matplotlib.pyplot as plt
import numpy as np

def volume(dimensions,epsilon):
    if epsilon < 1 : 
        v_peau = pow((1-2*epsilon), dimensions)
    else :
        v_peau = 0
    return v_peau

liste = []
liste_ratio = []
dimensions = []
epsilon = 0.1

for i in range(30):
    liste.append(volume(i,epsilon))
    liste_ratio.append(1-liste[i]/1)
    dimensions.append(i)
    
    
plt.plot(dimensions, liste_ratio)
    

```
