1. ´Ecrire une fonction LinearSearch(T,e) qui retourne l’indice de l’´element
e dans le tableau T s’il existe et -1 sinon.

`````python
import random as rd
import math

def LinearSearch(T,e):
    i = 0
    n = len(T)
    while i < n:
        if T[i] == e:
            return i
        else:
            i+= 1
    return -1
`````

2. ´Ecrire une fonction qui g´en´ere un tableau de n valeurs al´eatoires enti`eres

`````python
def gen_random_tab(n):
    T = []
    for _ in range(n):
        T.append(rd.randint(-10, 10))
    return T
`````

4. ´Ecrire une fonction DichoSearch(T,e) qui retourne l’indice de l’´element e
dans le tableau T s’il existe et -1 sinon par la m´ethode dichotomique.  

`````python
def DicoSearch(T,e):
    g,d = 1,len(T)
    while g <= d:
        m = (g+d)//2
        if T[m] < e:
            g = m+1
        elif T[m] > e:
            d = m-1
        else:
            return m
    return -1
`````

6. ´Ecrire une version r´ecursive de la recherche dichotomique.

`````python
def DicoSearch_recursive(T,e,g=0,d=None):
    if d is None:
        d = len(T)-1
    if g > d:
        return -1
    m = (g+d)//2
    if T[m]==e:
        return m
    elif T[m] < e:
        return DicoSearch_recursive(T,e,m+1,d)
    else:
        return DicoSearch_recursive(T,e,g,m-1)
`````

6. V´erifier que le nombre de comparaisons est en O(√n)


On cherche à minimiser : $f(k) = n/k + k$

En dérivant f(k) : $f'(k) = - n/k^2 +1$  
En résolvant f'(k)=0 :  
$n/k² +1 =0$ ==> $n/k²= 1 $ ==> k² = n

Donc la valeur optimale de k est k=$\sqrt n$ pour obtenir un équilibre entre le nombre de sauts et la taille de la recherche séquentielle.  

Finalement, on a montré que le nombre de comparaisons est $O(n/k + k$) or k=$\sqrt n $.  
Donc on a une complexité de O($\sqrt n$).
