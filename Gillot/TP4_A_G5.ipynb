### La méthode itérative

1. ´Ecrire une fonction it´erative Fibonacci_iterative(n) qui retourne le n-ième terme de la suite de Fibonacci.

````python
global steps
    
# Méthode itérative
def Fibonacci_iterative(n):
    global steps
    if n == 0:
        steps += 1
        return 0
    elif n == 1:
        steps += 1
        return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        steps += 1
        a, b = b, a + b
    return b
````

2. Compter le nombre d’´etapes.

````python
# Méthode de comptage d'étapes
def measure_steps(func,n):
    global steps
    steps = 0
    func(n)
    return steps


# Exemple 
print(measure_steps(Fibonacci_iterative,30))
````

- Cas particulier : Si n=0 ou n=-1 on aura 1 étape.
- Sinon on boucle de 2 à n inclus :   
A chaque itération, on incrémente de 1 le nombre d'étapes donc (n-1) étapes pour cette partie.  

Au final, on aura n étapes.

````python
# Méthode récursive
def Fibonacci_recursive(n):
    global steps
    steps += 1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return Fibonacci_recursive(n - 1) + Fibonacci_recursive(n - 2)

# Méthode récursive alternative
#def Fibonacci_recursive(n):
#    global steps
#    if n == 0:
#        steps += 1
#        return 0
#    elif n == 1:
#        steps += 1
#        return 1
#    steps += 1
#    return Fibonacci_recursive(n - 1) + Fibonacci_recursive(n - 2)
````

2. ´Ecrire une fonction mult_matrix(M1,M2) qui retourne le produit matriciel
de deux matrices 2 × 2. Une matrice M =
(a b
c d
)
est repr´esent´ee par la
liste [[a,b],[c,d]].


````python
# Multiplication de matrices 2x2
def mult_matrix(M1, M2):
    return [
        [M1[0][0]*M2[0][0] + M1[0][1]*M2[1][0],
         M1[0][0]*M2[0][1] + M1[0][1]*M2[1][1]],
        [M1[1][0]*M2[0][0] + M1[1][1]*M2[1][0],
         M1[1][0]*M2[0][1] + M1[1][1]*M2[1][1]]
    ]
````

3. ´Ecrire une fonction square_matrix(M) qui retourne le carr´e de la matrice M.

````python
# Carré d'une matrice 2x2
def square_matrix(M):
    return mult_matrix(M, M)
````

4. ´Ecrire une fonction pow_phi(n) qui retourne Φn en utilisant la m´ethode
“Square and Multiply” it´erative. Indication : l’´el´ement neutre pour la
multiplication des matrices est la matrice identit´e, ici Id =
(1 0
0 1)

````python
# Matrice de base Φ
Phi = [[1, 1], [1, 0]]

# Puissance de Φ (Square and Multiply)
def pow_phi(n):
    global steps
    result = [[1, 0], [0, 1]]  # matrice identité
    M = Phi[:]  # copie de Φ
    while n > 0:
        steps += 1
        if n % 2 == 1:
            result = mult_matrix(result, M)
        M = mult_matrix(M, M)
        n //= 2
    return result

print(pow_phi(1))  # [[1,1],[1,0]]
print(pow_phi(2))  # [[2,1],[1,1]]
print(pow_phi(3))  # [[3,2],[2,1]]
````

5. En d´eduire une fonction Fibonacci_matrix(n) qui retourne le n-ième terme de la suite de Fibonacci.

````python
# Méthode matricielle
def Fibonacci_matrix(n):
    if n == 0:
        return 0
    Mn = pow_phi(n - 1)  # Φ^(n-1)
    return Mn[0][0]      # Fn est en haut à gauche

for i in range(11):
    print(f"F({i}) =", Fibonacci_matrix(i))
````

6. Estimer la complexit´e du nombres d’´etapes de cette fonction

On appelle pow_phi(n-1) pour calculer $\Phi^{n-1} par exponentielle rapide.   
Pour pow_phi(n) utilise Square&Multiply :   
- On compare par la matrice identité.  
- Tant que n>0, On multiplie le résultat par la matrice courante ssi n est impaire et on la met au carré.  
- n = n // 2.  

Nombres d'itérations :  
La boucle while n>0 fait environ log_2(n) itérations.  
A chaque itération, on effectue une ou deux multiplications de matrices 2x2.  

Complexité d'une matrice 2x2 :   
Chaque produit de matrice 2x2 nécessite 4 produits et 2 additions. Donc cela est constant : O(1)  

Complexité totale :   
On fait O(log_n) itérations avec un coût constant par itérations. Donc on a un nombre d'étapes total de O(log_n).  

Donc la complexité est O(log_n).

### Comparaison des performances

````python
#Comparaison des Performances

n_values = [10, 20, 30, 35]

steps_values_iterative = []
steps_values_recursive = []
steps_values_matrix = []

for n in n_values:
    steps_values_iterative.append(measure_steps(Fibonacci_iterative,n))
    steps_values_recursive.append(measure_steps(Fibonacci_recursive,n))
    steps_values_matrix.append(measure_steps(Fibonacci_matrix,n))
    
print("Performances pour les valeurs [10, 20, 30, 35] (itératif, récursif et matriciel)")
print(steps_values_iterative,steps_values_recursive,steps_values_matrix)


n_values = [10**4,10**5,10**6]

steps_values_iterative = []
steps_values_matrix = []

for n in n_values:
    steps_values_iterative.append(measure_steps(Fibonacci_iterative,n))
    steps_values_matrix.append(measure_steps(Fibonacci_matrix,n))

print("\nPerformances pour les valeurs [10**4,10**5,10**6] (itératif et matriciel)")
print(steps_values_iterative,steps_values_recursive,steps_values_matrix)
````

Pour n=10 toutes les méthodes sont très rapides.
Pour n=20, la méthode récursive ralentit.
Pour n=30, la méthode récursive est très lente.
Cependant à partir de n>35, la fonction récursive devient inutilisable (très lente).

Pour plus, cela reste calculable mais le temps d'exécution devient bien plus long.
